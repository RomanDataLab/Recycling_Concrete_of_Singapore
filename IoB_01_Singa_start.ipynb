```python
#identifying lat-long point
# point = location.latlng
point = (1.2874,103.8525)
radius = 1000
```

**STREET NETWORK**

```python
#retreiving graph from that point in a range of 2000 m
geodata = ox.graph_from_point(point, dist=radius)
```

```python
#using osmnx library to plot the graph(matplotlib)
fig, ax = ox.plot_graph(geodata)
```

```python
#retrive nodes and edges
nodes, edges = ox.graph_to_gdfs(geodata)
```
